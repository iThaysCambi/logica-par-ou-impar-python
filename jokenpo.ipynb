```python
import random

# nomes dos jogadores
jogador1 = input("Digite o nome do Jogador 1: ")
modo = input("Jogar contra (1) outro jogador ou (2) computador? ")

if modo == "1":
    jogador2 = input("Digite o nome do Jogador 2: ")
else:
    jogador2 = "Computador"

pontuacao1 = 0
pontuacao2 = 0


def escolher_jogada():
    print('\n' + '='*20)
    print('Opções: pedra, papel, tesoura')
    print('='*20)

    movimento1 = input(f'{jogador1}, escolha: ').lower()

    if jogador2 == "Computador":
        movimento2 = random.choice(["pedra", "papel", "tesoura"])
        print(f'{jogador2} escolheu: {movimento2}')
    else:
        movimento2 = input(f'{jogador2}, escolha: ').lower()

    return movimento1, movimento2


def jogar(m1, m2):
    global pontuacao1, pontuacao2

    print(f'\n{jogador1} escolheu {m1}')
    print(f'{jogador2} escolheu {m2}')

    if m1 not in ["pedra", "papel", "tesoura"] or m2 not in ["pedra", "papel", "tesoura"]:
        print("❌ Jogada inválida!")
        return

    if m1 == m2:
        print("🤝 Empate!")

    elif (m1 == "pedra" and m2 == "tesoura") or \
         (m1 == "papel" and m2 == "pedra") or \
         (m1 == "tesoura" and m2 == "papel"):
        print(f"🏆 {jogador1} venceu!")
        pontuacao1 += 1

    else:
        print(f"🏆 {jogador2} venceu!")
        pontuacao2 += 1


# loop do jogo
while True:
    m1, m2 = escolher_jogada()
    jogar(m1, m2)

    print('\n📊 Placar:')
    print(f'{jogador1}: {pontuacao1}')
    print(f'{jogador2}: {pontuacao2}')

    continuar = input("\nDeseja jogar novamente? (s/n): ").lower()
    if continuar != 's':
        print("\n🎮 Fim de jogo!")
        break
```
